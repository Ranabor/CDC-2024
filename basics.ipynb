```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/Business_Dataset.csv')
airports = np.sort(pd.unique(df["airport_1"]))
airport_code = pd.read_csv('data/usairports.csv')
airport_code = airport_code[np.isin(airport_code["location_id"], airports)]
df.info()
#show all columns
pd.set_option('display.max_columns', None)
df.describe(include='all')
```

```python
flight_stats = df[["Year", "airport_2", "passengers"]]
flight_stats.head()
```

```python
grouped_flight_stats = flight_stats.groupby(["airport_2", "Year"]).sum().reset_index()
grouped_flight_stats
```

```python
airports = ["ATL", "DFW", "DEN", "LAX", "ORD", "JFK", "MCO", "LAS", "CLT", "MIA"]
colors = {
    'ATL': '#BF3B6A',  # Dark Pink
    'DFW': '#BF8B3B',  # Dark Gold
    'DEN': '#BFBF3B',  # Olive Green
    'LAX': '#3BBFB3',  # Dark Teal
    'ORD': '#3B6ABF',  # Dark Blue
    'JFK': '#6A3BBF',  # Dark Purple
    'MCO': '#BF3B3B',  # Dark Red
    'LAS': '#BF6A3B',  # Dark Orange
    'CLT': '#6ABF3B',  # Dark Lime Green
    'MIA': '#3BBF6A'   # Dark Mint
}

specific_aiports = grouped_flight_stats[np.isin(grouped_flight_stats["airport_2"], airports)]
specific_aiports = specific_aiports[(specific_aiports['Year'] > 1995) & ~np.isin(specific_aiports["Year"], [2020, 2021, 2024])]
specific_aiports = specific_aiports.pivot(index='Year', columns='airport_2', values='passengers').fillna(0)
plt.figure(figsize=(12, 6))
for airport in specific_aiports.columns:
    plt.plot(specific_aiports.index, specific_aiports[airport], marker='o', label=airport, color=colors.get(airport, 'black'))

plt.title('Number of Passengers by Airport Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.xticks(specific_aiports.index, rotation=45)
plt.legend(title='Airports')
plt.grid()
plt.tight_layout() 

plt.show()
```

```python
flight_stats_outbound = df[["Year", "airport_1", "passengers"]]
flight_stats_outbound.head()
grouped_flight_stats_outbound = flight_stats_outbound.groupby(["airport_1", "Year"]).sum().reset_index()

specific_aiports = grouped_flight_stats_outbound[np.isin(grouped_flight_stats_outbound["airport_1"], airports)]
specific_aiports = specific_aiports[(specific_aiports['Year'] > 1995) & ~np.isin(specific_aiports["Year"], [2020, 2021, 2024])]
specific_aiports = specific_aiports.pivot(index='Year', columns='airport_1', values='passengers').fillna(0)
plt.figure(figsize=(12, 6))
for airport in specific_aiports.columns:
    plt.plot(specific_aiports.index, specific_aiports[airport], marker='o', label=airport, color=colors.get(airport, 'black'))

plt.title('Number of Passengers by Airport Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.xticks(specific_aiports.index, rotation=45)
plt.legend(title='Airports')
plt.grid()
plt.tight_layout() 

plt.show()
```
