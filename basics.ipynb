```python
flight_stats = df[["Year", "airport_2", "passengers"]]
flight_stats.head()
```

```python
grouped_flight_stats = flight_stats.groupby(["airport_2", "Year"]).sum().reset_index()
grouped_flight_stats
```

```python
airports = ["ATL", "DFW", "DEN", "LAX", "ORD", "JFK", "MCO", "LAS", "CLT", "MIA", "BOS", "FLL"]
colors = {
    'ATL': '#BF3B6A',  # Dark Pink
    'DFW': '#BF8B3B',  # Dark Gold
    'DEN': '#BFBF3B',  # Olive Green
    'LAX': '#3BBFB3',  # Dark Teal
    'ORD': '#3B6ABF',  # Dark Blue
    'JFK': '#6A3BBF',  # Dark Purple
    'MCO': '#BF3B3B',  # Dark Red
    'LAS': '#BF6A3B',  # Dark Orange
    'CLT': '#6ABF3B',  # Dark Lime Green
    'MIA': '#3BBF6A'   # Dark Mint
}

specific_aiports = grouped_flight_stats[np.isin(grouped_flight_stats["airport_2"], airports)]
specific_aiports = specific_aiports[(specific_aiports['Year'] > 1995) & ~np.isin(specific_aiports["Year"], [2020, 2021, 2024])]
specific_aiports = specific_aiports.pivot(index='Year', columns='airport_2', values='passengers').fillna(0)
plt.figure(figsize=(12, 6))
for airport in specific_aiports.columns:
    plt.plot(specific_aiports.index, specific_aiports[airport], marker='o', label=airport, color=colors.get(airport, 'black'))

plt.title('Number of Passengers by Airport Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.xticks(specific_aiports.index, rotation=45)
plt.legend(title='Airports')
plt.grid()
plt.tight_layout() 

plt.show()
```

```python
miles_vs_fare = df["nsmiles"] / df["fare"]
plt.hist(miles_vs_fare, bins=np.arange(0, 20, 0.5))
plt.axvline(np.mean(miles_vs_fare), color='red', linestyle='dashed', linewidth=2, label='Mean: {:.2f}'.format(np.mean(miles_vs_fare)))
plt.axvline(np.median(miles_vs_fare), color='green', linestyle='dashed', linewidth=2, label='Median: {:.2f}'.format(np.median(miles_vs_fare)));
```

```python
def get_mileage_airport(airport_name: str, year: int):
    temp = df[(df["airport_1"] == airport_name) & (df["Year"] == year)]
    miles_vs_fare = temp["nsmiles"] / temp["fare"]
    plt.hist(miles_vs_fare, bins=np.arange(0, 20, 0.5), alpha=0.6, label=airport_name + " " + str(year))
    plt.legend()
    # plt.axvline(np.mean(miles_vs_fare), color='red', linestyle='dashed', linewidth=2, label='Mean: {:.2f}'.format(np.mean(miles_vs_fare)))
    # plt.axvline(np.median(miles_vs_fare), color='green', linestyle='dashed', linewidth=2, label='Median: {:.2f}'.format(np.median(miles_vs_fare)))
```

```python
df_relevant = df[(df["Year"] > 1995) & (df["Year"] != 2024)]

def generate_traffic_graph(airports, start_year, end_year, removed_years = [2020, 2021, 2024], inbound = True):
    direction = "airport_2" if inbound else "airport_1"
    direction_word = "Inbound" if inbound else "Outbound"
    chosen_airports = df_relevant[np.isin(df_relevant[direction], airports)]
    chosen_airports["passengers"] *= 365 / 1_000_000
    chosen_airports = chosen_airports[np.isin(chosen_airports["Year"], np.arange(start_year, end_year))]
    chosen_airports = chosen_airports[~np.isin(chosen_airports["Year"], removed_years)]
    grouped_flights = chosen_airports.groupby([direction, "Year"]).sum().reset_index()
    grouped_flights = grouped_flights.pivot(index="Year", columns=direction, values="passengers").fillna(0)
    
    plt.figure(figsize=(12, 6))
    for airport in grouped_flights.columns:
        plt.plot(grouped_flights.index, grouped_flights[airport], marker='o', label=airport)

    plt.title("Number of " + direction_word + " Passengers by Airport Over the Years")
    plt.xlabel('Year')
    plt.ylabel('Number of Passengers (in Millions)')
    plt.xticks(grouped_flights.index, rotation=45)
    plt.legend(title='Airports')
    plt.grid()
    plt.tight_layout() 

    plt.show()

def get_top_10_airports(inbound = True):
    direction = "airport_2" if inbound else "airport_1"
    summed_airports = df_relevant[[direction, "passengers"]].groupby(direction).sum().reset_index()
    summed_airports = summed_airports.sort_values(by=["passengers"], ascending=False)
    return list(summed_airports[direction][:10])

top_inbound_airports = get_top_10_airports()
top_outbound_airports = get_top_10_airports(False)

top_inbound_airports, top_outbound_airports
```

```python
generate_traffic_graph(top_inbound_airports, 1996, 2024)
```

```python
generate_traffic_graph(top_outbound_airports, 1996, 2024, inbound=False)
```

```python
airport_colors = {
        "ATL": {"inbound": "#1f77b4", "outbound": "#aec7e8"},  # Blue shades
        "DFW": {"inbound": "#ff7f0e", "outbound": "#ffbb78"},  # Orange shades
        "DEN": {"inbound": "#2ca02c", "outbound": "#98df8a"},  # Green shades
        "LAX": {"inbound": "#d62728", "outbound": "#ff9896"},  # Red shades
        "ORD": {"inbound": "#9467bd", "outbound": "#c5b0d5"},  # Purple shades
    }

def generate_comparison(airports, start_year, end_year, removed_years = [2020, 2021, 2024], inbound = True):

    direction = "airport_2"
    chosen_airports = df_relevant[np.isin(df_relevant[direction], airports)]
    chosen_airports["passengers"] *= 365 / 1_000_000
    chosen_airports = chosen_airports[np.isin(chosen_airports["Year"], np.arange(start_year, end_year))]
    chosen_airports = chosen_airports[~np.isin(chosen_airports["Year"], removed_years)]
    grouped_flights = chosen_airports.groupby([direction, "Year"]).sum().reset_index()
    grouped_flights = grouped_flights.pivot(index="Year", columns=direction, values="passengers").fillna(0)

    direction = "airport_1"
    chosen_airports = df_relevant[np.isin(df_relevant[direction], airports)]
    chosen_airports["passengers"] *= 365 / 1_000_000
    chosen_airports = chosen_airports[np.isin(chosen_airports["Year"], np.arange(start_year, end_year))]
    chosen_airports = chosen_airports[~np.isin(chosen_airports["Year"], removed_years)]
    grouped_flights_2 = chosen_airports.groupby([direction, "Year"]).sum().reset_index()
    grouped_flights_2 = grouped_flights_2.pivot(index="Year", columns=direction, values="passengers").fillna(0)

    
    plt.figure(figsize=(12, 6))
    for airport in grouped_flights.columns:
        plt.plot(grouped_flights.index, grouped_flights[airport], marker='o', color=airport_colors[airport]["inbound"], label=f"{airport} (Inbound)")
        plt.plot(grouped_flights_2.index, grouped_flights_2[airport], marker='o', linestyle='--', color=airport_colors[airport]["outbound"], label=f"{airport} (Outbound)")

    plt.title("International Airports Inbound vs Outbound")
    plt.xlabel('Year')
    plt.ylabel('Number of Passengers (in Millions)')
    plt.xticks(grouped_flights.index, rotation=45)
    plt.legend(title='Airports')
    plt.grid()
    plt.tight_layout() 

    plt.show()

generate_comparison(["ATL", "DFW", "DEN", "LAX", "ORD"], 1996, 2024)
```
